# Patent mini project

***Problem Statement :*** Give number of sub-patents corresponding to each patent ID. Read data from 'patent' file.

***Variables reference :***
    
    - text : contains data extracted from the 'patent' file
    
    - ans_dict : contains required ans. Keys represent patent ID and corresponding value is a list of sub patents under that ID

```python
with open('patent') as f: ##reading data from patent file
    text = f.read().split('\n')
```

```python
text = [s.split() for s in text] ## seaprating patent ID and sub-patents
```

```python
text = text[:-1] ## stripping empty list at the end
```

```python
patents = [int(text[i][0]) for i in range(len(text))] ## list of patents(duplicates allowed)
```

```python
sub_patents = [float(text[i][1]) for i in range(len(text))] ##list of sub-patents
```

```python
sub_patents
```

```python
ans_dict = {} ##contains the answer
ans_dict = {k:[] for k in range(1,1000)}
ans_dict
```

```python
for i in sub_patents:
    p = int(i)
    ans_dict[p].append(i)
```

```python
ans_dict
```

```python
for k,v in ans_dict.items():
    print(k, len(v))
```
