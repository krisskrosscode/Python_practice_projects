# Temperature mini project

***Problem Statement :*** Give the maximum temperature for each year. Read data from 'temperature' file.

***Variables reference :***
    
    - text : contains data extracted from the 'temperature' file
    
    - ans_dict : contains required ans. Keys represent year and corresponding values represent maximum temperature in that year

```python
with open('Temperature') as f: ## reading the data from file and storing it into 'text' variable
    text = f.readlines()
```

```python
text = [s.strip('\n') for s in text] ##removing unnecessary spaces and newline characters
```

```python
text = text[1:] ## removing the header data ['year', 'temp'], Now we have pure data
```

```python
text
```

```python
text = [s.split() for s in text]    ##separating year and temperature data
```

```python
text
```

```python
ans_dict = {} ## ans will be stored in this dictionary
```

```python

for i in text:
    m = int(i[1])
    if int(i[0]) in ans_dict:
        if m > ans_dict[int(i[0])]:
            ans_dict[int(i[0])] = m
            
    else:
        ans_dict[int(i[0])] = m
```

```python
print('year', 'maximum temperature') ##printing required answer
for year, maxtemp in ans_dict.items():
    print(year, maxtemp)
```
